```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'data/CdSe_CdZnS Core_Shell.txt'
with open(path, 'r') as spectrum:
    spectrum = spectrum.readlines()

print(spectrum)
freq = []
val = []

for line in spectrum:
    for i in range(2):
        id = line.find(',')
        line = line.replace(',', '.')

    freq.append(float(line.split('\t')[0]))
    val.append(float(line[:-1].split('\t')[1]))

val = np.array(val)
freq = np.array(freq)
```

```python
mean = 0
quadratic = 0
for el in val:
    mean += el

mean = mean/len(val)

for el in val:
    quadratic += (el - mean)**2

quadratic = np.sqrt(quadratic/len(val))

print(mean, quadratic)
```
